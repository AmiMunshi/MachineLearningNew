```python
#import the relevant libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split 
from sklearn.preprocessing import Normalizer
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_squared_error
from sklearn.tree import DecisionTreeClassifier 
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
df= pd.read_csv('/content/diabetes.csv') #read the dataset
```

```python
#to see if any columns has empty cells
for column in df:
  print(" Number of empty cells in {} is {} ". format(column, (df[column]=="").sum()))
```

```python
#To check how many zeros are there in each column
for column in df:
  print(("Number of zeros in {} is {}".format(column,(df[column]==0).sum())))
```

```python
from sklearn.impute import SimpleImputer
impute = SimpleImputer(missing_values=0,strategy='mean')

df[['Glucose','BloodPressure','SkinThickness','Insulin','BMI']]=impute.fit_transform(df[['Glucose','BloodPressure','SkinThickness','Insulin','BMI']])
```

```python
df.describe()
```

```python
df.head()
```

```python
#To check how many zeros are there in each column
for column in df:
  print(("Number of zeros in {} is {}".format(column,(df[column]==0).sum())))
```

```python
#Define the independent and dependent variables
y= df['Outcome'] #dependent variable is price
x= df.drop(['Outcome'], axis=1)
```

```python
# splitting the data
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size= 0.3,random_state=1)
```

```python
#Implementing DecisionTree using sklearn
model = DecisionTreeClassifier()
model.fit(x_train,y_train)
```

```python
#Make prediction for the test data
y_pred= model.predict(x_test) 
```

```python
ConfusionMatrix = confusion_matrix(y_test, y_pred)
print(ConfusionMatrix)
```

```python
#Accuracy score
Accuracy=accuracy_score(y_test,y_pred)
print("Accuracy of the model is- ", Accuracy)
```
